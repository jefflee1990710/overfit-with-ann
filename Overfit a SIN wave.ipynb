```python
import keras
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_df = pd.DataFrame(np.random.sample([10000, 1]) * 10, columns=['X'])
train_df['y'] = train_df['X'].apply(lambda x : math.sin(x))
```

```python
plt.scatter(train_df['X'], train_df['y'])
```

```python
def get_model():
    model_input = keras.Input(shape=(1,))
    x = keras.layers.Dense(6, activation='relu')(model_input)
    x = keras.layers.Dense(1, activation='sigmoid')(x)
    model = keras.Model(model_input, x)
    model.compile(loss='mean_squared_error', optimizer=keras.optimizers.Adam(lr=0.1))
    return model

model = get_model()
model.summary()
```

```python
def show_compare():
    train_df['pred'] = model.predict(train_df['X'])
    plt.figure(figsize=(12, 6))
    ax = plt.subplot()
    ax.scatter(train_df['X'], train_df['y'])
    ax.scatter(train_df['X'], train_df['pred'])
```

```python
model.fit(train_df['X'], train_df['y'], epochs=500)
show_compare()
```
